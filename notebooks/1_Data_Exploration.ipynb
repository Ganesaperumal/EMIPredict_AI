Welcome to the EMIPredict AI project! Run the cells below as instructed.

```python
import os
import pandas as pd
import numpy as np
df = pd.read_csv('../data/raw/emi_prediction_dataset.csv')
```

```python
df.info()
df.isnull().sum()
```

```python
categorical_cols = df.select_dtypes(include=['object', 'string']).columns

for col in categorical_cols:
    print(df[col].value_counts(),"\n---------------")
```

### age

```python
# df[~df['age'].astype(str).str.match(r'^\d+$')]['age'].value_counts(dropna=False)
print(df['age'].unique())
print("Null-Count:", df['age'].isna().sum())
```

```python
df['age'] = pd.to_numeric(df['age'].astype(str).str.split('.').str[0], errors='coerce')
print(df['age'].value_counts(dropna=False))
print("\nNull-Count:", df['age'].isna().sum())
```

### gender

```python
print(df['gender'].value_counts(dropna=False))
print("\nNull-Count:", df['gender'].isna().sum())
```

```python
df['gender'] = df['gender'].str[0].str.upper().map({'M': 'Male', 'F': 'Female'})
print(df['gender'].value_counts(dropna=False))
print("\nNull-Count:", df['gender'].isna().sum())
```

### education

```python
print(df['education'].value_counts(dropna=False))
print("\nNull-Count:", df['education'].isna().sum())
```

```python
df['education'] = df['education'].replace(['nan', 'NaN'], df['education'].mode()[0]).fillna(df['education'].mode()[0])

print(df['education'].value_counts(dropna=False))
print("\nNull-Count:", df['education'].isna().sum())
```

### monthly_salary

```python
unique_monthly_salary = df['monthly_salary'].unique()
print("Random unique abnormal values:\n", np.random.choice(unique_monthly_salary, size=min(100, len(unique_monthly_salary)), replace=False))

print("\nNull-Count:", df['monthly_salary'].isna().sum())
```

```python
df['monthly_salary'] = pd.to_numeric(df['monthly_salary'].astype(str).str.split('.').str[0], errors='coerce')
print(df['monthly_salary'].value_counts(dropna=False))
print("\nNull-Count:", df['monthly_salary'].isna().sum())
```

### monthly_rent

```python
print(df['monthly_rent'].describe())
print("\nNull-Count:", df['monthly_rent'].isna().sum())
```

```python
df['monthly_rent'] = df['monthly_rent'].fillna(df['monthly_rent'].median())
print("\nNull-Count:", df['monthly_rent'].isna().sum())
```

### bank_balance

```python
unique_bank_balance = df['bank_balance'].unique()
print("Random unique abnormal values:\n", np.random.choice(unique_bank_balance, size=min(100, len(unique_bank_balance)), replace=False))

print("\nNull-Count:", df['bank_balance'].isna().sum())
```

```python
df['bank_balance'] = pd.to_numeric(df['bank_balance'].astype(str).str.split('.').str[0], errors='coerce')
print(df['bank_balance'].value_counts(dropna=False))
print("\nNull-Count:", df['bank_balance'].isna().sum())
```

```python
df['bank_balance'] = df['bank_balance'].fillna(df['bank_balance'].median())
print("\nNull-Count:", df['bank_balance'].isna().sum())
```

### credit_score

```python
print(df['credit_score'].describe())
print("\nNull-Count:", df['credit_score'].isna().sum())
```

```python
df['credit_score'] = df['credit_score'].fillna(df['credit_score'].median())
print("\nNull-Count:", df['credit_score'].isna().sum())
```

### emergency_fund

```python
print(df['emergency_fund'].describe())
print("\nNull-Count:", df['emergency_fund'].isna().sum())
```

```python
df['emergency_fund'] = df['emergency_fund'].fillna(df['emergency_fund'].median())
print("\nNull-Count:", df['emergency_fund'].isna().sum())
```
